```python
import numpy as np
import pandas as pd
from sklearn import datasets
import matplotlib.pyplot as plt

%matplotlib inline

import seaborn as sns
```

```python
from urllib.request import urlretrieve
url = 'https://www4.stat.ncsu.edu/~boos/var.select/diabetes.tab.txt'
urlretrieve(url, 'diabetes.tab.txt')
```

```python
df = pd.read_csv('diabetes.tab.txt',sep='\t')
df
```

```python
df.rename(columns={'S1':'T_cells', 'S2':'ldl', 'S3':'hdl', 'S4':'tch', 'S5':'ltg', 'S6':'glu'},inplace=True)
df
```

# matrix plots

### heatmaps
## [correlation](https://realpython.com/numpy-scipy-pandas-correlation-python/)
  
#Correlation Between x and y  
#equal to 1	perfect positive linear relationship  
#greater than 0	positive correlation  
#equal to 0	independent  
#less than 0	negative correlation  
#equal to -1	perfect negative linear relationship  

```python
df_matrix=df.corr() #will be used for making heat map 
df_matrix #show how each feature influences others
```

```python
plt.figure(figsize=(30,8))
sns.set_context('paper',font_scale=1.4)

sns.heatmap(df_matrix, annot=True, cmap='Blues')
                
```

```python
df_pivot=df.pivot_table(index='AGE',columns='SEX',values='BMI')
df_pivot
```

```python
sns.heatmap(df_pivot.T, cmap='Blues',linecolor='green',linewidth=1)
```
