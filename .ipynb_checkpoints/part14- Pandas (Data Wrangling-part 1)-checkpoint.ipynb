###  note: this errors are because of fact due to
### 1. apply() on series applies a funtion to values
###  2. But apply() of Dataframes applies funtion to underlying series not onto the values.
### for each value of DataFrame we use "applymap"

```python
def female_population(x): #x is whole dataframe that is passed in
    female_population= x["Gender_Ratio"]*1000*x["Population"]/(x["Gender_Ratio"]*1000+1000)
    male_population= x["Population"]-female_population
    return female_population #two returns will take single cell therefore two seprate funtions
def male_population(x):
    female_population= x["Gender_Ratio"]*1000*x["Population"]/(x["Gender_Ratio"]*1000+1000)
    male_population= x["Population"]-female_population
    return male_population
```

```python
mf_apply=pd.DataFrame(columns=['male','female'])
mf_apply
```

```python
mf_apply['female']=copy.apply(female_population,axis=1)
mf_apply['male']=copy.apply(male_population,axis=1)
mf_apply=mf_apply.astype('int') #converts float datatype to integers
mf_apply
```

##### why 2 functions for separte male and female popoulation?
#####  both return values are stored in same cell so, we can skip hasel to separte them!

```python
def population(x):
    female_population= x["Gender_Ratio"]*1000*x["Population"]/(x["Gender_Ratio"]*1000+1000)
    male_population= x["Population"]-female_population
    return male_population,female_population #two returns from functions

mf_population=pd.DataFrame(columns=['male','female'])
mf_population
```

```python
mf_population['male']=copy.apply(population,axis=1)
mf_population
```

### more efficiently!! we can write production ready functional programs

```python
mf_apply=pd.DataFrame(columns=['male','female'])
mf_apply
```

```python
global count;
count=0;
def mf_population(x):
    global count
    female_population= x["Gender_Ratio"]*1000*x["Population"]/(x["Gender_Ratio"]*1000+1000)
    male_population= x["Population"]-female_population;
    if count==1:
        return female_population;
    if count==0:
        return male_population;
    
def call():
    return copy.apply(mf_population,axis=1);


def mf():
    global count
    if count==1:
        mf_apply['female']=call()
        count=1;
    if count==0:
        mf_apply['male']=call()
        count=1;
    return;

for x in range(2):
    mf()
    
    
mf_apply=mf_apply.astype('int')
mf_apply
```
