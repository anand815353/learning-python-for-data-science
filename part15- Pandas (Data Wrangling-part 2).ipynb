```python
import time
```

perf_counter() function always returns the float value of time in seconds. Return the value (in fractional seconds) of a performance counter, i.e. a clock with the highest available resolution to measure a short duration. It does include time elapsed during sleep and is system-wide.

```python
df = pd.DataFrame(np.random.randint(0, 10, size=(100, 4)), columns=list('ABCD'))
df
```

### 2.1 Using Index

```python
for index in df.index: #uses index to iterate
    print(index)
```

```python
def loop_with_index(df):
    tic=time.perf_counter() #Records start time
    temp=0
    for index in df.index:  #uses index to iterate through dataframe
        temp += df['A'][index] + df['B'][index] #uses index to sum up all the elemnts in dataframe
    print("sum=",temp)
    return time.perf_counter()-tic  # returns total time taken

tock=loop_with_index(df)
tock
```

### 2.2 Using loc

```python
len(df)
```

```python
for index in range(len(df)): #uses index
        print(index)
```

```python
def loop_with_loc(df):
    tic=time.perf_counter() #Records start time
    temp=0
    for index in range(len(df)):
        temp += df.loc[index,'A'] + df.loc[index,'B'] #uses loc to sum up all the elemnts in dataframe
                                                      # but needs index to change row one by one
    print("sum=",temp)
    return time.perf_counter()-tic  # returns total time taken

tock=loop_with_loc(df)
tock
```

### 2.3 Using iloc

```python
len(df)
```

```python
for index in range(len(df)): #uses index
        print(index)
```

```python
def loop_with_iloc(df): #same as loc but insted of label it needs integer index
    tic=time.perf_counter()
    temp=0
    for index in range(len(df)):
        temp += df['A'].iloc[index] + df['B'].iloc[index]  #series is selected before iterating(diifrent synatx of iloc)
    print("sum=",temp)
    return time.perf_counter()-tic

tock=loop_with_iloc(df)
tock
```

### 2.4 iterrows()

```python
for index,row in df.iterrows(): #uses this to iterate
    print(index,row)
```

```python
def loop_with_iterrows(df):
    tic=time.perf_counter()
    temp=0
    for index,row in df.iterrows():
        temp += row.A + row.B  
    print("sum=",temp)
    return time.perf_counter()-tic

tock=loop_with_iterrows(df)
tock
```

### 2.5 itertuples()

```python
for row_tuple in df.itertuples(): #uses tuples to iterate
    print(row_tuple)
```

```python
def loop_with_itertuples(df):
    tic=time.perf_counter()
    temp=0
    for row_tuple in df.itertuples():
        temp += row_tuple.A + row_tuple.B  
    print("sum=",temp)
    return time.perf_counter()-tic

tock=loop_with_itertuples(df)
tock
```

### 2.6 iteritems()

```python
for labels,contents in df.iteritems(): #uses this to iterate
    print(labels,contents)
```

```python
def loop_with_iteritems1(df):  #it iterates over one column at a time 
    tic=time.perf_counter()
    temp=0
    for labels,contents in df.iteritems():
        if (labels == 'A') or (labels == 'B'):
            temp=temp+sum(contents)
        else: pass
    print("sum=",temp)
    return time.perf_counter()-tic

tock=loop_with_iteritems1(df)
tock
```

```python
def loop_with_iteritems2(df):  #it iterates over one column at a time 
    tic=time.perf_counter()
    temp=0
    for labels,contents in df.iteritems():
        if (labels == 'A') or (labels == 'B'):
            temp=temp+sum(contents)
        else:
            continue
    print("sum=",temp)
    return time.perf_counter()-tic

tock=loop_with_iteritems2(df)
tock
```

### 2.7 iteration over columns

```python
for labels in df: #uses labels to iterate
    print(labels)
```

```python
def custom_loop_over_columns(df):
    tic=time.perf_counter()
    temp=0
    for labels in df:
        if (labels == 'A') or (labels == 'B'):
            for i in df.index:
                temp=temp+df.loc[i,labels]
    print("sum=",temp)
    return time.perf_counter()-tic

tock=custom_loop_over_columns(df)
tock
```

### 2.8 iteration using apply

```python
def loop_using_apply(df):
    tic=time.perf_counter()
    temp=df.apply(lambda x: x['A'] + x['B'], axis=1).sum()
    print("sum=",temp)
    return time.perf_counter()-tic

tock=loop_using_apply(df)
tock
```

### 2.9 iteration using zip

```python
for a, b in zip(df['A'], df['B']): 
    print(a,b)
```

```python
def loop_with_zip(df):
    tic=time.perf_counter()
    temp = 0
    for a, b in zip(df['A'], df['B']):
        temp += a + b
    print("sum=",temp)
    return time.perf_counter()-tic

tock=loop_with_zip(df)
tock
```

### 2.10 iteration using pandas builtin function

```python
print(df['A'],df['B'])
```

```python
def loop_with_pandas_builtin(df):
    tic=time.perf_counter()
    (df['A'] + df['B']).sum()
    return time.perf_counter()-tic

tock=loop_with_pandas_builtin(df)
tock
```

### 2.11 iteration using numpy builtin function

```python
print(df['A'].values, df['B'].values)
```

```python
def loop_with_numpy_builtin(df):
    tic=time.perf_counter()
    temp=(df['A'].values + df['B'].values).sum()
    print("sum=",temp)
    return time.perf_counter()-tic

tock=loop_with_numpy_builtin(df)
tock
```

# 3. time-comparision of iteration methods of dataframes
##### Will compare diffrent methods based on time taken by each method on diffrent size of dataframe
##### to determine which method is suitable for huge dataset.

### logic

```python
# logic of loop to make dataframes for testing time taken by each iteration method in "2.Iterating over rows and columns."

no_of_iterations=[3,5,8]
for i in no_of_iterations: #this loop will make datframe 1st of size 3x2 then 5x2 and finaly 8x2
    df1 = pd.DataFrame(np.random.randint(0, 10, size=(i, 2)), columns=list('AB')) 
    
    # functions will be called here one by one
    #            ""
    #            ""
    #            ""
    
    print(i,"iterations\n",df1.T,"\n\n")  #prints transpose version of dataframe... i.e it creates datframes of diffrent size
```

### setup

```python
#row_index and will also be used as size of dataFrames for test
no_of_iteration=[100,1000,5000,10000,50000,100000,500000,1000000,5000000,10000000,50000000,100000000,1000000000,10000000000]


#column_names
itr=['Using Index','loc','iloc','iterrows','itertuples','iteritems','over columns','apply','zip','pandas builtin','numpy builtin']
```

```python
time_comparision=pd.DataFrame(index=no_of_iteration,columns=itr) #creating  dataframe to store time taken by each method
time_comparision
```

#### row 100 will store time taken by each method to iterate on dataframe of size 100 
#### row 1000 "____________________________________________________" size 1000
### and so on...

### the TEST

```python
#this loop will automate the creation of dataframes of multiple sizes and will call each function in "2.Iterating over rows and columns."

for i in no_of_iteration: 
    df1 = pd.DataFrame(np.random.randint(0, 10, size=(i, 2)), columns=list('AB'))
    
    time_comparision.loc[i,'Using Index'] = loop_with_index(df1)
    time_comparision.loc[i,'loc'] = loop_with_loc(df1)
    time_comparision.loc[i,'iloc'] = loop_with_iloc(df1)
    time_comparision.loc[i,'iterrows'] = loop_with_iterrows(df1)
    time_comparision.loc[i,'itertuples'] = loop_with_itertuples(df1)
    time_comparision.loc[i,'iteritems'] = loop_with_iteritems1(df1)
    time_comparision.loc[i,'over columns'] = custom_loop_over_columns(df1)
    time_comparision.loc[i,'apply'] = loop_using_apply(df1)
    time_comparision.loc[i,'zip'] = loop_with_zip(df1)
    time_comparision.loc[i,'pandas builtin'] = loop_with_pandas_builtin(df1)
    time_comparision.loc[i,'numpy builtin'] = loop_with_numpy_builtin(df1)


time_comparision  # beyond 100000000 will take lot of time therefor skipped!!!  but data should be sufficient to infer...
```
