##  Two interface:(1. MATLAB style interface, 2. Object oriented interface)

###  MATLAB style interface 
    • object hierarchy structure
    • The MATLAB-style tools are contained in the pyplot (plt) interface. 
    • this interface is stateful: it keeps track of the “current” figure and axes, which are where all plt commands are applied. 

```python
plt.figure()                    # create a plot figure

# create the first of two panels and set current axis; 
#subplot(row, column, cell no) 

plt.subplot(2, 1, 1)            # subplot(2, 1, 1) <-- break subplot in 2 x 1 grid and place subplot in 1st cell.
plt.plot(x, np.sin(x))

# create the second panel and set current axis

plt.subplot(2, 1, 2)           #subplot(2, 1, 2) <-- break subplot in 2 x 1 grid and place subplot in 2nd cell.
plt.plot(x, np.cos(x));     
```

#### Disadvantage: we cannot go back to 1st panel after creating 2nd panel.

```python
fig1 = plt.subplot(1,3,1)    # here subplot divides the grid in 1 row and 2 column
fig1.plot(x,np.sin(x))
fig1 = plt.subplot(1,3,2) 
fig1.plot(x,np.cos(x));
```

```python
fig1 = plt.subplot(1,3,3) 
fig1.plot(x,np.tan(x));
```

```python
fig1 = plt.subplot(3,1,1)    #therefore should be same notebook cell.
fig1.plot(x,np.sin(x))
fig1 = plt.subplot(3,1,2) 
fig1.plot(x,np.cos(x));
fig1 = plt.subplot(3,1,3) 
fig1.plot(x,np.tan(x));
```

### Object-oriented interface 
    • More control over the figure.
    • we create a figure object and call it later.
    • object-oriented interface the plotting functions are methods of explicit Figure and Axes objects.
    • A Figure object can contain one or more axes objects. 
    • One axes represents one plot inside figure.

### [subplot vs subplots](https://stackoverflow.com/questions/52214776/python-matplotlib-differences-between-subplot-and-subplots)

```python
fig = plt.figure()    # created Figure object its empty right now.
print(type(fig))      # type() used to find out the type of an object.
```

```python
fig.suptitle("Empty figure")    # fig.suptitle <--- Add a centered title to the figure.
plt.show()                      #no figure to show 
```

```python
ax = fig.add_subplot(1,1,1)   # adding axes in the fig (creates a grid of of subplots)
ax.set_title("Empty plot")    # Set the title of plot
plt.show()                    #no figure to show
```

```python
fig = plt.figure()

ax1 = fig.add_subplot(2,2,1)         # Generates a grid of 2x2 subplots and get axes object for 1st location
ax1.set_title('First Location')

ax2 = fig.add_subplot(2,2,2)         # Get the axes object for subplot at 2nd location
ax2.set_title('Second Location')

ax3 = fig.add_subplot(2,2,3)         # Get the axes object for subplot at 3rd location
ax3.set_xlabel('Third Location')

ax4 = fig.add_subplot(2,2,4)         # Get the axes object for subplot at 4th location
ax4.set_xlabel('Fourth Location')

plt.show()
```

##### multiple subplots in a figure

```python
fig,ax = plt.subplots(2)   #subplots has two return values canvas(fig) and its objects(ax)
ax[0].plot(x,np.sin(x))    #ax[0] <--- cell 1 
ax[1].plot(x,np.cos(x));   #ax[1] <--- cell 2
```

```python
# defining the values of X 
a =[0, 1, 2, 3, 4, 5, 6] 
  
# defining the value of Y 
b =[0, 1, 3, 6, 9, 12, 17] 
  
# creating the canvas with class 'fig' 
# and it's object 'axes' with '1' row  
# and '2' columns 
fig, axes1 = plt.subplots(2, 2) 
  
# plotting graph for 1st element 
axes1[0, 0].plot(a, b, 'g--o')     #g <---green
  
# plotting graph for 2nd element 
axes1[0, 1].plot(b, a, 'm--o')     
  
# plotting graph for 3rd element 
axes1[1, 0].plot(a, b, 'b--o')      #b <---blue
  
# plotting graph for 4th element 
axes1[1, 1].plot(b, a, 'r--o')      #r <---Red
```

## Simple Line Plots

```python
plt.plot(x, np.sin(x - 0), color='blue')           # specify color by name
plt.plot(x, np.sin(x - 1), color='g')              # short color code (rgbcmyk)
plt.plot(x, np.sin(x - 2), color='0.75')           # Grayscale between 0 and 1
plt.plot(x, np.sin(x - 3), color='#FFDD44')        # Hex code (RRGGBB from 00 to FF)
plt.plot(x, np.sin(x - 4), color=(1.0,0.2,0.3))    # RGB tuple, values 0 and 1
plt.plot(x, np.sin(x - 5), color='chartreuse');    # all HTML color names supported
```

***If no color is specified, Matplotlib will automatically cycle through a set of default
colors for multiple lines

```python
plt.plot(x, x + 0, linestyle='solid')
plt.plot(x, x + 1, linestyle='dashed')
plt.plot(x, x + 2, linestyle='dashdot')
plt.plot(x, x + 3, linestyle='dotted');
```

```python
plt.plot(x, x + 4, linestyle='-') # solid
plt.plot(x, x + 5, linestyle='--') # dashed
plt.plot(x, x + 6, linestyle='-.') # dashdot
plt.plot(x, x + 7, linestyle=':'); # dotted
```

```python
plt.plot(x, x + 0, '-g') # solid green
plt.plot(x, x + 1, '--c') # dashed cyan
plt.plot(x, x + 2, '-.k') # dashdot black
plt.plot(x, x + 3, ':r'); # dotted red
```

### Adjusting the Plot: Axes Limits

```python
plt.plot(x, np.sin(x))
plt.xlim(-1, 11)         #controls X-axis
plt.ylim(-1.5, 1.5);     #controls Y-axis
```

```python
# axis to be displayed in reverse
plt.plot(x, np.sin(x))
plt.xlim(10, 0)               # from 10 to 0
plt.ylim(1.2, -1.2);          # from +v 1.2 to -v 1.2
```

##### The plt.axis() method allows you to set the x and y limits with a single call, by passing a list that specifies [xmin, xmax, ymin, ymax]


##### note: here the potential confusion between axes with an e, and axis with an i.

```python
plt.plot(x, np.sin(x))
plt.axis([-1, 11, -1.5, 1.5]);
```

##### plt.axis() method auto‐matically tighten the bounds around the current plot.

```python
plt.plot(x, np.sin(x))
plt.axis('tight');
```

##### It allows even higher-level specifications, such as ensuring an equal aspect ratio so that on your screen, one unit in x is equal to one unit in y.

```python
plt.plot(x, np.sin(x))
plt.axis('equal');
```

### Labeling Plots

```python
plt.plot(x, np.sin(x))
plt.title("A Sine Curve")      #Title of the plot
plt.xlabel("x")                # x-axis label
plt.ylabel("sin(x)")           # Y-axis label
```

```python
plt.plot(x, np.sin(x), '-g', label='sin(x)')
plt.plot(x, np.cos(x), ':b', label='cos(x)')
plt.axis('equal')
plt.legend()                                 #Displayes the legend.
```

##### In the object-oriented interface to plotting, rather than calling these functions individually, it is often more convenient to use the ax.set() method to set all these properties at once.

```python
ax = plt.axes()
ax.plot(x, np.sin(x))
ax.set(xlim=(0, 10), ylim=(-2, 2),xlabel='x', ylabel='sin(x)',title='A Simple Plot');
```
